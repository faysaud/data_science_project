### Calculate the time elapsed in closed and creation date

```python
df['Closed Date']= pd.to_datetime(df['Closed Date'])
df['Created Date']=pd.to_datetime(df['Created Date'])
```

```python
df['Incident time elapsed']= df['Closed Date']-df['Created Date']
```

### Convert the calculated date to seconds to get a better representation

```python
df['Incident time elapsed']=df['Incident time elapsed'].dt.total_seconds()
```
